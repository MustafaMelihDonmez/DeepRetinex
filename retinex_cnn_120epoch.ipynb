```python
import os
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt
import pandas as pd

# --- Constants ---
EPS = 1e-6
MIN_SIGMA = 1.0
MAX_SIGMA = 30.0
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- Dataset ---
class LOLDataset(Dataset):
    def __init__(self, low_dir, high_dir, transform=None):
        self.low_images = sorted(os.listdir(low_dir))
        self.high_images = sorted(os.listdir(high_dir))
        self.low_dir = low_dir
        self.high_dir = high_dir
        self.transform = transform

    def __len__(self):
        return len(self.low_images)

    def __getitem__(self, idx):
        low_path = os.path.join(self.low_dir, self.low_images[idx])
        high_path = os.path.join(self.high_dir, self.high_images[idx])
        low = Image.open(low_path).convert("RGB")
        high = Image.open(high_path).convert("RGB")
        if self.transform:
            low = self.transform(low)
            high = self.transform(high)
        return low, high

# --- Retinex functions ---
def retinex_single_scale(img, sigma):
    img = img.float().clamp(min=EPS)
    log_img = torch.log(img)
    ksize = int(2 * sigma) * 2 + 1
    if ksize % 2 == 0:
        ksize += 1
    gaussian_kernel = torch.zeros((3, 1, ksize, ksize), device=img.device)
    for c in range(3):
        x = torch.linspace(-sigma, sigma, steps=ksize)
        kernel_1d = torch.exp(-0.5 * (x / sigma) ** 2)
        kernel_1d /= kernel_1d.sum()
        kernel_2d = torch.outer(kernel_1d, kernel_1d)
        kernel_2d /= kernel_2d.sum()
        gaussian_kernel[c, 0] = kernel_2d
    pad = ksize // 2
    img_padded = F.pad(img, (pad, pad, pad, pad), mode='reflect')
    result = torch.zeros_like(img)
    for c in range(3):
        result[:, c:c+1] = F.conv2d(img_padded[:, c:c+1], gaussian_kernel[c:c+1])
    log_gaussian = torch.log(result.clamp(min=EPS))
    return log_img - log_gaussian

def retinex_multi_scale(img, sigma_list):
    return sum(retinex_single_scale(img, s) for s in sigma_list) / len(sigma_list)

def apply_color_restoration(img, msr, alpha=125, beta=46, G=192):
    img_sum = torch.sum(img, dim=1, keepdim=True).clamp(min=EPS)
    cr = beta * (torch.log(alpha * img.clamp(min=EPS)) - torch.log(img_sum))
    return G * (msr + cr)

def normalize_retinex_output(output):
    bmin = output.view(output.size(0), -1).min(1)[0].view(-1,1,1,1)
    bmax = output.view(output.size(0), -1).max(1)[0].view(-1,1,1,1)
    return (output - bmin) / (bmax - bmin + EPS)

# --- Model ---
class RetinexCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 32, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(32, 64, 3, 2, 1), nn.ReLU(),
            nn.Conv2d(64, 128, 3, 2, 1), nn.ReLU(),
            nn.Conv2d(128, 256, 3, 2, 1), nn.ReLU()
        )
        self.gap = nn.AdaptiveAvgPool2d(1)
        self.sigma_predictor = nn.Sequential(
            nn.Flatten(),
            nn.Linear(256, 128), nn.ReLU(),
            nn.Linear(128, 3), nn.Softplus()
        )
        self.color_param_predictor = nn.Sequential(
            nn.Flatten(),
            nn.Linear(256, 128), nn.ReLU(),
            nn.Linear(128, 3), nn.Softplus()
        )
        self.enhancement_net = nn.Sequential(
            nn.Conv2d(3, 64, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(64, 64, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(64, 3, 3, 1, 1), nn.Sigmoid()
        )

    def forward(self, x):
        features = self.encoder(x)
        global_feat = self.gap(features)
        sigma_vals = 5.0 + 25.0 * self.sigma_predictor(global_feat)
        sigma_vals = sigma_vals.clamp(MIN_SIGMA, MAX_SIGMA)
        color_params = self.color_param_predictor(global_feat)
        alpha = 100.0 + 50.0 * color_params[:, 0:1]
        beta = 40.0 + 20.0 * color_params[:, 1:2]
        G = 150.0 + 100.0 * color_params[:, 2:3]
        msr_output = torch.zeros_like(x)
        for i in range(x.size(0)):
            img = x[i:i+1]
            sigmas = [sigma_vals[i, j].item() for j in range(3)]
            msr = retinex_multi_scale(img, sigmas)
            msrcr = apply_color_restoration(img, msr, alpha[i].item(), beta[i].item(), G[i].item())
            msr_output[i:i+1] = normalize_retinex_output(msrcr)
        return self.enhancement_net(msr_output)

# --- Loss and Accuracy ---
def ssim_loss(x, y):
    C1 = 0.01 ** 2
    C2 = 0.03 ** 2
    mu_x = F.avg_pool2d(x, 3, 1, 1)
    mu_y = F.avg_pool2d(y, 3, 1, 1)
    sigma_x = F.avg_pool2d(x * x, 3, 1, 1) - mu_x ** 2
    sigma_y = F.avg_pool2d(y * y, 3, 1, 1) - mu_y ** 2
    sigma_xy = F.avg_pool2d(x * y, 3, 1, 1) - mu_x * mu_y
    ssim_n = (2 * mu_x * mu_y + C1) * (2 * sigma_xy + C2)
    ssim_d = (mu_x**2 + mu_y**2 + C1) * (sigma_x + sigma_y + C2)
    return 1 - (ssim_n / ssim_d).mean()

def combined_loss(pred, target):
    return 0.5 * F.l1_loss(pred, target) + 0.5 * ssim_loss(pred, target)

def compute_ssim_accuracy(pred, target):
    return 1.0 - ssim_loss(pred, target).item()

# --- Training ---
def train_epoch(model, loader, opt):
    model.train()
    total_loss = 0
    total_ssim = 0
    for x, y in tqdm(loader, desc="Training"):
        x, y = x.to(DEVICE), y.to(DEVICE)
        out = model(x)
        loss = combined_loss(out, y)
        opt.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=5.0)
        opt.step()
        total_loss += loss.item()
        total_ssim += compute_ssim_accuracy(out, y)
    return total_loss / len(loader), total_ssim / len(loader)

# --- Main ---
def main():
    transform = transforms.Compose([
        transforms.Resize((256, 256)),
        transforms.ToTensor()
    ])
    LOW_DIR = "/content/lol-dataset/lol_dataset/our485/low"
    HIGH_DIR = "/content/lol-dataset/lol_dataset/our485/high"
    dataset = LOLDataset(LOW_DIR, HIGH_DIR, transform=transform)
    loader = DataLoader(dataset, batch_size=8, shuffle=True)

    model = RetinexCNN().to(DEVICE)
    opt = optim.Adam(model.parameters(), lr=5e-5)
    save_path = "/content/best_retinex_model.pth"

    best_loss = float('inf')
    ssim_scores = []
    loss_scores = []

    for epoch in range(120):
        print(f"Epoch {epoch+1}")
        loss, ssim_acc = train_epoch(model, loader, opt)
        print(f"Avg Loss: {loss:.4f} | Avg SSIM Accuracy: {ssim_acc:.4f}")
        loss_scores.append(loss)
        ssim_scores.append(ssim_acc)
        if loss < best_loss:
            best_loss = loss
            torch.save(model.state_dict(), save_path)
            print(f"Best model saved at epoch {epoch+1} with loss {loss:.4f}")

    # Plot and Save Results
    epochs = list(range(1, len(ssim_scores) + 1))
    plt.figure(figsize=(10, 4))
    plt.plot(epochs, ssim_scores, marker='o', label="SSIM")
    plt.xlabel("Epoch")
    plt.ylabel("SSIM")
    plt.title("CNN: SSIM Over Epochs")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig("/content/cnn_ssim_plot.png")
    plt.show()

    plt.figure(figsize=(10, 4))
    plt.plot(epochs, loss_scores, marker='o', label="Loss", color='red')
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("CNN: Loss Over Epochs")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig("/content/cnn_loss_plot.png")
    plt.show()

    df = pd.DataFrame({
        "Epoch": epochs,
        "CNN Loss": loss_scores,
        "CNN SSIM": ssim_scores
    })
    df.to_csv("/content/cnn_results.csv", index=False)
    print("Exported: cnn_results.csv, cnn_ssim_plot.png, cnn_loss_plot.png")

if __name__ == "__main__":
    main()
```

```python
import os
import torch
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms

# Load your trained CNN model
model = RetinexCNN().to(DEVICE)
model.load_state_dict(torch.load("/content/best_retinex_model.pth", map_location=DEVICE))
model.eval()

# Image transformation
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
])

# Choose 5 image filenames (without path)
image_names = ["10.png", "18.png", "24.png", "40.png", "12.png"]

# Prepare figure
plt.figure(figsize=(15, len(image_names) * 3))

# Loop through images
for i, name in enumerate(image_names):
    low_path = f"/content/lol-dataset/lol_dataset/our485/low/{name}"
    high_path = f"/content/lol-dataset/lol_dataset/our485/high/{name}"

    low_pil = Image.open(low_path).convert("RGB")
    low_tensor = transform(low_pil).unsqueeze(0).to(DEVICE)

    with torch.no_grad():
        enhanced_tensor = model(low_tensor).clamp(0, 1)

    def tensor_to_image(t):
        t = t.squeeze().detach().cpu()
        return transforms.ToPILImage()(t)

    low_img = tensor_to_image(low_tensor)
    enhanced_img = tensor_to_image(enhanced_tensor)

    high_img = None
    if os.path.exists(high_path):
        high_img = Image.open(high_path).convert("RGB").resize((256, 256))

    # Plot each row of 3 images
    plt.subplot(len(image_names), 3, 3 * i + 1)
    plt.imshow(low_img)
    plt.title(f"Input: {name}")
    plt.axis("off")

    plt.subplot(len(image_names), 3, 3 * i + 2)
    plt.imshow(enhanced_img)
    plt.title("Enhanced")
    plt.axis("off")

    plt.subplot(len(image_names), 3, 3 * i + 3)
    if high_img:
        plt.imshow(high_img)
        plt.title("Ground Truth")
    else:
        plt.text(0.5, 0.5, "No Ground Truth", ha="center", va="center")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
