```python
import os
import cv2
import numpy as np
from PIL import Image
from tqdm import tqdm
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt
import pandas as pd

# --- Constants ---
EPS = 1e-6
MIN_SIGMA = 1.0
MAX_SIGMA = 30.0
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- Dataset ---
class LOLDataset(Dataset):
    def __init__(self, low_dir, high_dir, transform=None):
        self.low_images = sorted(os.listdir(low_dir))
        self.high_images = sorted(os.listdir(high_dir))
        self.low_dir = low_dir
        self.high_dir = high_dir
        self.transform = transform

    def __len__(self):
        return len(self.low_images)

    def __getitem__(self, idx):
        low_path = os.path.join(self.low_dir, self.low_images[idx])
        high_path = os.path.join(self.high_dir, self.high_images[idx])
        low = Image.open(low_path).convert("RGB")
        high = Image.open(high_path).convert("RGB")
        if self.transform:
            low = self.transform(low)
            high = self.transform(high)
        return low, high

# --- Retinex Functions ---
def retinex_single_scale(img, sigma):
    img = img.float().clamp(min=EPS)
    log_img = torch.log(img)
    sigma = max(min(float(sigma), MAX_SIGMA), MIN_SIGMA)
    ksize = int(2 * sigma) * 2 + 1
    if ksize % 2 == 0:
        ksize += 1
    gaussian_kernel = torch.zeros((3, 1, ksize, ksize), device=img.device)
    for c in range(3):
        x = torch.linspace(-sigma, sigma, steps=ksize)
        kernel_1d = torch.exp(-0.5 * (x / sigma) ** 2)
        kernel_1d /= kernel_1d.sum()
        kernel_2d = torch.outer(kernel_1d, kernel_1d)
        kernel_2d /= kernel_2d.sum()
        gaussian_kernel[c, 0] = kernel_2d
    pad = ksize // 2
    img_padded = F.pad(img, (pad, pad, pad, pad), mode='reflect')
    result = torch.zeros_like(img)
    for c in range(3):
        result[:, c:c+1] = F.conv2d(img_padded[:, c:c+1], gaussian_kernel[c:c+1])
    log_gaussian = torch.log(result.clamp(min=EPS))
    return log_img - log_gaussian

def retinex_multi_scale(img, sigma_list):
    return sum(retinex_single_scale(img, s) for s in sigma_list) / len(sigma_list)

def apply_color_restoration(img, msr, alpha=125, beta=46, G=192):
    img_sum = torch.sum(img, dim=1, keepdim=True).clamp(min=EPS)
    cr = beta * (torch.log(alpha * img.clamp(min=EPS)) - torch.log(img_sum))
    return G * (msr + cr)

def normalize_retinex_output(output):
    bmin = output.view(output.size(0), -1).min(1)[0].view(-1, 1, 1, 1)
    bmax = output.view(output.size(0), -1).max(1)[0].view(-1, 1, 1, 1)
    return (output - bmin) / (bmax - bmin + EPS)

# --- U-Net Blocks ---
class UNetBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, padding=1),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.conv(x)

# --- Retinex U-Net Model ---
class RetinexUNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.enc1 = UNetBlock(3, 64)
        self.pool1 = nn.MaxPool2d(2)
        self.enc2 = UNetBlock(64, 128)
        self.pool2 = nn.MaxPool2d(2)
        self.enc3 = UNetBlock(128, 256)
        self.pool3 = nn.MaxPool2d(2)
        self.enc4 = UNetBlock(256, 512)
        self.pool4 = nn.MaxPool2d(2)

        self.bottleneck = UNetBlock(512, 1024)

        self.up4 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.dec4 = UNetBlock(1024, 512)
        self.up3 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.dec3 = UNetBlock(512, 256)
        self.up2 = nn.ConvTranspose2d(256, 128, 2, 2)
        self.dec2 = UNetBlock(256, 128)
        self.up1 = nn.ConvTranspose2d(128, 64, 2, 2)
        self.dec1 = UNetBlock(128, 64)

        self.final = nn.Conv2d(64, 3, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        e1 = self.enc1(x)
        e2 = self.enc2(self.pool1(e1))
        e3 = self.enc3(self.pool2(e2))
        e4 = self.enc4(self.pool3(e3))
        b = self.bottleneck(self.pool4(e4))
        d4 = self.dec4(torch.cat([self.up4(b), e4], dim=1))
        d3 = self.dec3(torch.cat([self.up3(d4), e3], dim=1))
        d2 = self.dec2(torch.cat([self.up2(d3), e2], dim=1))
        d1 = self.dec1(torch.cat([self.up1(d2), e1], dim=1))
        return self.sigmoid(self.final(d1))

# --- Loss and Accuracy ---
def ssim_loss(x, y):
    C1 = 0.01 ** 2
    C2 = 0.03 ** 2
    mu_x = F.avg_pool2d(x, 3, 1, 1)
    mu_y = F.avg_pool2d(y, 3, 1, 1)
    sigma_x = F.avg_pool2d(x * x, 3, 1, 1) - mu_x ** 2
    sigma_y = F.avg_pool2d(y * y, 3, 1, 1) - mu_y ** 2
    sigma_xy = F.avg_pool2d(x * y, 3, 1, 1) - mu_x * mu_y
    ssim_n = (2 * mu_x * mu_y + C1) * (2 * sigma_xy + C2)
    ssim_d = (mu_x**2 + mu_y**2 + C1) * (sigma_x + sigma_y + C2)
    return 1 - (ssim_n / ssim_d).mean()

def combined_loss(pred, target):
    return 0.5 * F.l1_loss(pred, target) + 0.5 * ssim_loss(pred, target)

def compute_ssim_accuracy(pred, target):
    return 1.0 - ssim_loss(pred, target).item()

# --- Training ---
def train_epoch(model, loader, opt):
    model.train()
    total_loss = 0
    total_ssim = 0
    for x, y in tqdm(loader, desc="Training"):
        x, y = x.to(DEVICE), y.to(DEVICE)
        msr_output = torch.zeros_like(x)
        for i in range(x.size(0)):
            img = x[i:i+1]
            sigmas = [15, 80, 250]
            msr = retinex_multi_scale(img, sigmas)
            msrcr = apply_color_restoration(img, msr)
            msr_output[i:i+1] = normalize_retinex_output(msrcr)
        out = model(msr_output)
        loss = combined_loss(out, y)
        opt.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=5.0)
        opt.step()
        total_loss += loss.item()
        total_ssim += compute_ssim_accuracy(out, y)
    return total_loss / len(loader), total_ssim / len(loader)

# --- Main ---
def main():
    transform = transforms.Compose([
        transforms.Resize((256, 256)),
        transforms.ToTensor()
    ])
    LOW_DIR = "/content/lol-dataset/lol_dataset/our485/low"
    HIGH_DIR = "/content/lol-dataset/lol_dataset/our485/high"
    dataset = LOLDataset(LOW_DIR, HIGH_DIR, transform=transform)
    loader = DataLoader(dataset, batch_size=8, shuffle=True)

    model = RetinexUNet().to(DEVICE)
    opt = optim.Adam(model.parameters(), lr=5e-5)
    save_path = "/content/best_retinex_unet_model.pth"

    best_loss = float('inf')
    ssim_scores = []
    loss_scores = []

    for epoch in range(120):
        print(f"Epoch {epoch+1}")
        loss, ssim_acc = train_epoch(model, loader, opt)
        print(f"Avg Loss: {loss:.4f} | Avg SSIM Accuracy: {ssim_acc:.4f}")
        loss_scores.append(loss)
        ssim_scores.append(ssim_acc)
        if loss < best_loss:
            best_loss = loss
            torch.save(model.state_dict(), save_path)
            print(f"Best model saved at epoch {epoch+1} with loss {loss:.4f}")

    # --- Graphs and CSV ---
    epochs = list(range(1, len(ssim_scores) + 1))
    plt.figure(figsize=(10, 4))
    plt.plot(epochs, ssim_scores, marker='o', label="U-Net SSIM")
    plt.xlabel("Epoch")
    plt.ylabel("SSIM")
    plt.title("U-Net: SSIM Over Epochs")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig("/content/unet_ssim_plot.png")
    plt.show()

    plt.figure(figsize=(10, 4))
    plt.plot(epochs, loss_scores, marker='o', color='red', label="U-Net Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("U-Net: Loss Over Epochs")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig("/content/unet_loss_plot.png")
    plt.show()

    df = pd.DataFrame({
        "Epoch": epochs,
        "U-Net Loss": loss_scores,
        "U-Net SSIM": ssim_scores
    })
    df.to_csv("/content/unet_results.csv", index=False)
    print("Exported: unet_results.csv, unet_ssim_plot.png, unet_loss_plot.png")

if __name__ == "__main__":
    main()
```

```python
import os
import torch
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms

# Load the trained U-Net model
model = RetinexUNet().to(DEVICE)
model.load_state_dict(torch.load("/content/best_retinex_unet_model.pth", map_location=DEVICE))
model.eval()

# Transformation
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
])

# Choose 5 image filenames to test
image_names = ["10.png", "18.png", "24.png", "40.png", "12.png"]

# Display figure
plt.figure(figsize=(15, len(image_names) * 3))

# Process each image
for i, name in enumerate(image_names):
    low_path = f"/content/lol-dataset/lol_dataset/our485/low/{name}"
    high_path = f"/content/lol-dataset/lol_dataset/our485/high/{name}"

    low_pil = Image.open(low_path).convert("RGB")
    low_tensor = transform(low_pil).unsqueeze(0).to(DEVICE)

    with torch.no_grad():
        sigmas = [15, 80, 250]
        msr = retinex_multi_scale(low_tensor, sigmas)
        msrcr = apply_color_restoration(low_tensor, msr)
        retinex_input = normalize_retinex_output(msrcr)
        enhanced = model(retinex_input).clamp(0, 1)

    def tensor_to_image(t):
        t = t.squeeze().detach().cpu()
        return transforms.ToPILImage()(t)

    low_img = tensor_to_image(low_tensor)
    enhanced_img = tensor_to_image(enhanced)

    high_img = None
    if os.path.exists(high_path):
        high_img = Image.open(high_path).convert("RGB").resize((256, 256))

    # Plot each row of 3 images
    plt.subplot(len(image_names), 3, 3 * i + 1)
    plt.imshow(low_img)
    plt.title(f"Input: {name}")
    plt.axis("off")

    plt.subplot(len(image_names), 3, 3 * i + 2)
    plt.imshow(enhanced_img)
    plt.title("Enhanced (U-Net)")
    plt.axis("off")

    plt.subplot(len(image_names), 3, 3 * i + 3)
    if high_img:
        plt.imshow(high_img)
        plt.title("Ground Truth")
    else:
        plt.text(0.5, 0.5, "No Ground Truth", ha="center", va="center")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
